```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Dataset.csv")
df.head()
```

```python
# Checking dataset structure
df.info()

# Checking column names
df.columns
```

```python
# Count of different rating texts (e.g., Excellent, Good, Average)
plt.figure(figsize=(8,5))
sns.countplot(x='Rating text', data=df, palette="viridis", order=df['Rating text'].value_counts().index)
plt.title("Distribution of Review Categories")
plt.xlabel("Review Category")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
print("This shows how many restaurants fall under Excellent / Good / Average / Poor categories.")
```

```python
# Average Aggregate Rating per Review Category
plt.figure(figsize=(8,5))
sns.barplot(x='Rating text', y='Aggregate rating', data=df, palette="coolwarm",
            order=df['Rating text'].value_counts().index)
plt.title("Average Rating per Review Category")
plt.xlabel("Review Category")
plt.ylabel("Average Aggregate Rating")
plt.xticks(rotation=45)
plt.show()
print("This shows the average rating for each review text (reviews linked to numerical ratings.")
```

```python
# Distribution of Votes
plt.figure(figsize=(8,5))
sns.histplot(df['Votes'], bins=40, kde=True, color='purple')
plt.title("Distribution of Votes")
plt.xlabel("Number of Votes")
plt.ylabel("Frequency")
plt.show()
print("This shows if votes are concentrated at low numbers or spread out.")
```

```python
# Relationship between Votes and Ratings 
plt.figure(figsize=(10,6))
sns.regplot(x="Aggregate rating", y="Votes", data=df, scatter_kws={"alpha":0.5, "s":50}, 
            line_kws={"color":"red"}, color="green")
plt.title("Votes vs Aggregate Rating with Trend Line", fontsize=14, weight="bold")
plt.xlabel("Aggregate Rating", fontsize=12)
plt.ylabel("Number of Votes", fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
print("Helps us see if restaurants with higher ratings also receive more votes.")
```

```python
# Top 10 most voted restaurants
top_votes = df[['Restaurant Name','Votes']].sort_values(by="Votes", ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x="Votes", y="Restaurant Name", data=top_votes, palette="mako")
plt.title("Top 10 Most Voted Restaurants")
plt.xlabel("Number of Votes")
plt.ylabel("Restaurant Name")
plt.show()
print("Shows the most popular restaurants based on votes")
```

```python
from wordcloud import WordCloud, STOPWORDS

# Combine all review texts into one string
text = " ".join(str(review) for review in df['Rating text'].dropna())

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, 
                      background_color="white", 
                      stopwords=STOPWORDS, 
                      colormap="viridis").generate(text)

# Plot the word cloud
plt.figure(figsize=(12,6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Restaurant Review Texts")
plt.show()
print("The word cloud shows the most common review terms. Positive keywords (like 'Excellent' or 'Very Good') dominate, indicating overall customer satisfaction trends.")
```

```python
# Selecting numeric features
corr_features = df[['Aggregate rating', 'Votes', 'Price range']]

# Compute correlation
corr = corr_features.corr()

# Plot heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation between Ratings, Votes, and Price Range")
plt.show()
print("The heatmap shows a positive correlation between ratings and votes, meaning highly rated restaurants get more customer engagement. Price range has a weaker correlation, suggesting affordability is not the only factor driving votes.")
```
